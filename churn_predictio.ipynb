```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold
cv=RepeatedStratifiedKFold(n_splits=5,n_repeats=2,random_state=1)
param={
    "n_estimators":[100,200,300,400,500],
    "criterion":["gini","entropy","log_loss"],
    "min_samples_split":[0.1,0.3,0.5],
    "max_depth":[1,2,3,4]
}
model_rfc=RandomForestClassifier()
search=RandomizedSearchCV(model_rfc, param, n_iter=10, scoring='accuracy', n_jobs=-1, cv=cv, random_state=1,verbose=1)
model=search.fit(x_train,y_train)
```

```python
model.best_score_
```

```python
model.best_params_
```

```python
model.best_estimator_
```

```python
classification_model=model.best_estimator_
classification_model2=classification_model.fit(x_train,y_train)
classification_model2
```

```python
prediction1=classification_model2.predict(x_train)

prediction1
```

```python
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
p=accuracy_score(prediction1,y_train)
p
```

```python
from sklearn.metrics import classification_report
print(classification_report(prediction1,y_train))
```
