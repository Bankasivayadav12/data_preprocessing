```python
#Univariate Analysis
#1.Histogram
#Here we’ll be performing univariate analysis on Numerical variables 
#using the histogram function.
#import warnings
#warnings.filterwarnings("ignore")
for i in data.select_dtypes(include="int").columns:
    sns.histplot(data=data,x=i,kde=True)
    plt.show()

#sns.histplot(data['age'],kde=True)
```

```python
#2.Bar Chart
#Univariate analysis of categorical data. We’ll be using the count 
#plot function from the seaborn library
sns.countplot(y=data['gender_full'])
```

```python
sns.barplot(data['gender_full'])
```

```python
#box plot
for i in data.select_dtypes(include="number").columns:
    sns.boxplot(data=data,y=i)
    plt.show()
#sns.boxplot(data['EmployeeID'])
```

```python
#pie-chart:A piechart helps us to visualize the percentage of the
#data belonging to each category
x = data['STATUS_YEAR'].value_counts() 
plt.pie(x.values, labels=x.index, autopct='%1.1f%%') 
plt.show() 
```

```python
'''
Bivariate analysis:Bivariate analysis is the simultaneous analysis 
of two variables. It explores the concept of the relationship 
between two variable whether there exists an association and the 
strength of this association or whether there are differences between
two variables and the significance of these differences.'''
'''
The main three types we will see here are:
1.Categorical v/s Numerical 
2.Numerical V/s Numerical
3.Categorical V/s Categorical data
'''
```

```python
#1.Categorical v/s Numerical
#department_name v/s length_of_service
plt.figure(figsize=(15, 5)) 
sns.barplot(x=data['department_name'],y=data['length_of_service']) 
plt.xticks(rotation='vertical') 
```

```python
#Numerical v/s Numerical
#scatterplot
#sns.scatterplot(x=data['length_of_service'],y=data['age']) 
#data.plot(x = 'length_of_service', y = 'age', kind = 'scatter');
#plt.show() 
for i in ['EmployeeID', 'age', 'length_of_service', 'store_name']:
    sns.scatterplot(data=data,x=i,y='STATUS_YEAR')
    plt.show()

#It displays the age and length of service of employees in the 
#organization as we can see that younger employees have 
#less experience in terms of their length of service.
```

```python
#multivariate:It is an extension of bivariate analysis which means 
#it involves multiple variables at the same time to find correlation 
#between them. Multivariate Analysis is a set of statistical model
#that examine patterns in multidimensional data by considering at
#once, several data variable.

from sklearn import datasets, decomposition 
iris = datasets.load_iris() 
X = iris.data 
y = iris.target 
pca = decomposition.PCA(n_components=2) 
X = pca.fit_transform(X) 
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y) 
```

```python
#HeatMap
#Here we are using a heat map to check the correlation between all
#the columns in the dataset. It is a data visualisation technique 
#that shows the magnitude of the phenomenon as colour in two 
#dimensions. The values of correlation can vary from -1 to 1
# where -1 means strong negative and +1 means strong positive correlation.
sns.heatmap(data=data.select_dtype(include=[np.int64,np.float64]).corr(),annot=True)
```
